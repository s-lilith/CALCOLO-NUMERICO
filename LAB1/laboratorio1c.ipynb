sys.float_info è una tupla che contiene informazione riguardo il tipo float, quali informazioni di basso livell riguardo la precisione e la rappresentazione interna.

max: rappresenta il massimo numero float finito rappresentabile che è 1.7976931348623157e+308, tutti i numeri al di sopra di questo vengono definiti infiniti (inf)

max_exp: rappresenta il massimo numero intero "e" tale che radix**(e-1) sia rappresentabile, cioè 1024

max_10_exp: rappresenta il massimo numero intero "e" tale che 10**e sia rappresentabole, cioè 308

```python
# Write a code to compute the machine precision ε in (float) default precision with a while construct. Compute also the mantissa digits number.
print(np.finfo(float).eps)
eps = 1
while eps + 1 > 1:
    eps /= 2
    print(eps)
eps *= 2
print("The machine epsilon is:", eps)
```

```python
# help(np.float16)
eps=1
fleps=np.float16(eps)
fl16= []
while fleps + 1 > 1:
    fleps /= 2
    fl16.append(fleps)
    print("%.40f" %fleps)
fleps *= 2
print("The machine epsilon is:", fleps)
```

```python
# help(np.float32)
eps=1
fleps=np.float32(eps)
fl32= []
while fleps + 1 > 1:
    fleps /= 2
    fl32.append(fleps)
    print("%.30f" %fleps)
fleps *= 2
print("The machine epsilon is:", fleps)
```

```python
headers= ["float16", "float32"]
data=[[fl16], [fl32]]
pandas.DataFrame(fl16, fl32, columns= ['float16________float32'])
```

2.2 Fibonacci and approximation

• Write a script that, given an input number n, computes the numbers of
the Fibonacci sequence that are less than n.



```python
def fibonacci(n):
    if n<2:
        return [1,1]
    else:
        L= fibonacci(n-1)
        if L[-1] < n:
            L.append(L[-1] + L[-2])
        return L

print(fibonacci(90))
```
